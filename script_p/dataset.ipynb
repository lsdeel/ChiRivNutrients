```python
stats_yr=ntrs.groupby(['省份','站点名称']).agg({'监测时间':['min','max']}).reset_index()
```

```python
stats_yr=stats_yr.rename(columns={'站点名称':'断面'})
```

```python
stats.merge(stats_yr,on=['省份','断面']).to_csv(wd+'/output/tab/nation_wq_stations_dateRange.csv',encoding='ANSI')
```
